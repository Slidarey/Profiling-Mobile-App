## Duplicate data

### Part 1 - Affirming data duplicates
The Google Play Store dataset has some duplicate data. An example of duplicated data will be printed for confirmation.

```python
for each_app in google_data:
   if each_app[0] == 'Instagram':
        print(each_app)
```

Now, we can see that the Instagram app is duplicated four times in the Google Play dataset, lets see how many duplicates we have in total.

```python
google_duplicate_apps = []
google_unique_apps = []

for each_app in google_data:
    if each_app[0] in google_unique_apps:
        google_duplicate_apps.append(each_app[0])
    else:
        google_unique_apps.append(each_app[0])
        
print('Number of duplicated rows is: '+ str(len(google_duplicate_apps)))
```

As seen, we have 1181 duplicated apps. Let's see how many times each app was duplicated.

```python
google_dupl_dict = {}

for each_app in google_duplicate_apps:
    if each_app in google_dupl_dict:
        google_dupl_dict[each_app] += 1
    else:
        google_dupl_dict[each_app] = 1
        
print(google_dupl_dict)
```

This shows the number of times each app was duplicated. We need to remove this duplicates based on the number of reviews, that is, the duplicate with the highest number of reviews will be selected and others will be removed. This criterion is so because the duplicate with the highest number of reviews signifies the most recent data for that particular app.

The next step is to select the app entry with the highest number of reviews.

### Part 2 - Removing data duplicates

We created a dictionary that contains each unique app with their maximum number of reviews

```python
google_reviews_max = {}

for each_app in google_data:
    name = each_app[0]
    n_reviews = float(each_app[3])
    
    if name in google_reviews_max:
        google_reviews_max[name] = max(google_reviews_max[name], n_reviews)
    else:
        google_reviews_max[name] = n_reviews
        
print(len(google_reviews_max))
```

This shows that the we 9659 unique apps in the google dataset.

The unique apps with highest number of reviews in the google play store dataset are extracted to the `android_clean` list  and the `already_added` list is used to avoid duplicating apps that have more than one entry with same number of reviews

```python
google_cleaned_data = []
already_added = []

for each_app in google_data:
    name = each_app[0]
    n_reviews = float(each_app[3])
    
    if (n_reviews == google_reviews_max[name]) and (name not in already_added):
        google_cleaned_data.append(each_app)
        already_added.append(each_app[0])
```

Exploring the cleaned google dataset and confirming that the number of rows is 9,659

```python
explore_data(google_cleaned_data, 0, 3, True)
```

The output shows that we have 9,659 rows as expected.



## Removing Non-English Apps

### Part 1
We need to define a function that will be used to test if the name of an app is english or not since or targeted audience is english-speaking.

But first, let's test the function out with some examples:

```python
def check_if_english(app_name):
    for each_charac in app_name:
        charac_num = ord(each_charac)
        if charac_num > 127:
            return False
        
    return True
print(check_if_english('Instagram'))
print(check_if_english('a奇艺PPS -《欢乐颂2》电视剧热播'))
print(check_if_english('Docs To Go™ Free Office Suite'))
print(check_if_english('Instachat 😜'))
```

Apparently, the `check_if_english` function seems to differentiate between a non-english and an english name, but some english names contain characters that are not within the ASCII range. Using this function as it is will make us remove useful data therefore, we need to modify the function.


### Part 2

We modify the `check_if_english` function to remove apps whose names have more than 3 characters whose corresponding number is out of ASCII range. Though, this function is not perfect but it will reduce the impact of data loss.

```python
def check_if_english(app_name):
    flag = 0
    for each_charac in app_name:
        charac_num = ord(each_charac)
        if charac_num > 127:
            flag += 1
            if flag > 3:
                return False
        
    return True

print(check_if_english('Instagram'))
print(check_if_english('a奇艺PPS -《欢乐颂2》电视剧热播'))
print(check_if_english('Docs To Go™ Free Office Suite'))
print(check_if_english('Instachat 😜'))
```

Filtering both datasets

```python
google_english = []
apple_english = []

for each_app in google_cleaned_data:
    if check_if_english(each_app[0]):
        google_english.append(each_app)

for each_app in apple_data:
    if check_if_english(each_app[2]):
        apple_english.append(each_app)
        
explore_data(google_english, 0, 3, True)
explore_data(apple_english, 0, 3, True)
```

Firstly, we will need to develop a frequency table to identify the most common genre for the two markets. This means we need frequency table for the `prime_genre` column for App Store dataset, and `Genres` and `Category` column for Google Play Store dataset.

### Part 2

We will need to create two functions to better represent the frequency of each genre for the two datasets:
    1. A function that generates the frequency table and shows the value for each genre in percentage
    2. Another function that exhibits the percentages in descending order.

```python
def freq_table(dataset, index):     # First function to generate frequency tables
    column_dict = {}
    total = 0
    
    for each_element in dataset:
        if each_element[index] in column_dict:
            column_dict[each_element[index]] += 1
        else:
            column_dict[each_element[index]] = 1
            
    for each_element in column_dict:
        total += column_dict[each_element]
    
    for each_element in column_dict:
        column_dict[each_element] = column_dict[each_element] / total * 100
        
    return column_dict
```

```python
def display_table(dataset, index):     # Second functin to sort the freq table in descending order
    table = freq_table(dataset, index)
    table_tuples = []
    
    for key in table:
        key_tuple = (table[key], key)
        table_tuples.append(key_tuple)
        
    sorted_table = sorted(table_tuples, reverse = True)
    
    for each_entry in sorted_table:
        print(each_entry[1], ':', each_entry[0])
```

## Most Popular Apps by Genre on AppStore

To find out the most popular app by genre we need to calculate the average number of installs for each app genre. This information can be found in the Google Play Store dataset under the `Install` column. In the App Store dataset, we don't have this information, but we can use the number of user ratings as a proxy, which can be found in the `rating_count_tot` column.

### Part 1

To calculate the number of installs per app genre for the App Store, we need to:
    1. Isolate the apps of each genre
    2. Find the sum of the user ratings for each genre
    3. Divide the sum by the number apps in that genre

Let's define a function to carry out the 3 aforementioned steps,

```python
def installs_per_genre(freq_table, main_dataset, genre_index, installs_index): # function to show installs per app genre
    installs_table = []
    for genre in freq_table:
        total = 0
        len_genre = 0
        
        for app in main_dataset:
            genre_app = app[genre_index]
            if genre_app == genre:
                app_usr_ratings = float(app[installs_index])
                total += app_usr_ratings
                len_genre += 1
        
        avg_no_users = total / len_genre
        
        installs_per_genre_tuple = (avg_no_users, genre)
        installs_table.append(installs_per_genre_tuple)
        
    sorted_table = sorted(installs_table, reverse = True)
    
    for each_tuple in sorted_table:
        print(each_tuple[1], ':', each_tuple[0])
```

Let's generate the frequency table of App Store and show the number of installs per app gnere using the `installs_per_genre` function,

```python
apple_app_genre = freq_table(apple_final, 12)
installs_per_genre(apple_app_genre, apple_final, 12, 6)
```

## Most Popular App By Genre on the Google Play Store


For the google dataset, we have data about the number of installs, with this, we should be able to get aclearer picture of the most popular app by genre. However, the values under the `install` column are open-ended(100+, 1,000+, 5,000+ etc). For example, we don't know if an with 100+ installs have 101, 200 or 500 installs, but we don't really need precise data for our analysis, we only want to know which app genres attract most installs, so we will leave values as they are taht is, we will assume that an app with 100+ installs have 100 installs.

To perform our computation however, we need to convert the values of the installs to float; to do this, we need to remove the commas `,` and the plus `+` character from the install values using the `str.replace(old, new)` function.

```python
google_app_genre = freq_table(google_final, 1)

install_freq_table = []

for category in google_app_genre:
    total = 0
    len_category = 0
    
    for each_app in google_final:
        category_app = each_app[1]
        if category_app == category:
            cat_string = each_app[5].replace('+', '')
            cat_string = cat_string.replace(',', '')
            
            cat_float = float(cat_string)
            
            total += cat_float
            len_category +=1
    
    avg_no_user = total / len_category
    
    cat_tuple = (avg_no_user, category)
    install_freq_table.append(cat_tuple)
    
srted_install_table = sorted(install_freq_table, reverse = True)

for each_cat in srted_install_table:
    print(each_cat[1], each_cat[0])
```

We can see that the `COMMUNICATION` category has the most number of installs; but we can't just assume it to be the best category for our app, we need to check the distribution of users for each categories.
First, let's sort the Google Play Store dataset with respect to `installs` for easy analysis.

```python
sorted_google_final = []
installs_for_google = []

for app in google_final:
    cat_str = app[5].replace('+', '')
    cat_str = cat_str.replace(',', '')
    cat_flt = float(cat_str)
    app_tuple = (cat_flt, app[0])
    installs_for_google.append(app_tuple)
srtd_google_installs = sorted(installs_for_google, reverse = True)

for each_tuple in srtd_google_installs:
    
    for each_app in google_final:
        
        cat_str_com = each_app[5].replace('+', '')
        cat_str_com = cat_str_com.replace(',', '')
        cat_flt_com = float(cat_str_com)
        
        if(each_app[0] == each_tuple[1]) and (cat_flt_com == each_tuple[0]):
            sorted_google_final.append(each_app)
```
